## 01. Load Dataset

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
iris.target_names
```

```python
iris.feature_names
```

## 02. Create Pandas Dataframe

```python
import pandas as pd

data = pd.DataFrame({'sepal length': iris.data[:,0],
                     'sepal width': iris.data[:,1],
                     'petal length': iris.data[:,2],
                     'petal width': iris.data[:,3],
                     'species': iris.target})
```

```python
data.head()
```

## 03. Divide Data into X and Y

```python
x = data[['sepal length', 'sepal width', 'petal length', 'petal width']]
x
```

```python
y = data['species']
y
```

## 04. Spliting data in Training and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.3)
```

```python
len(xTrain)
```

## 05. Select The Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100, criterion='gini') # n_estimators means no of decision trees 
```

## 06. Train the Model

```python
model.fit(xTrain, yTrain)
```

## 07. Predict the Model

```python
model.predict(xTest)
```

```python
# yTest
```

## 08. Check Accuracy

```python
model.score(xTest, yTest)
```
