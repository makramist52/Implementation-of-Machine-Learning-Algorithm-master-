#### Import Libraries

```python
import pandas as pd
from sklearn import linear_model
```

#### Read Dataset

```python
df = pd.read_csv('Insurance.csv')
df
```

#### Replace the NaN value in height column with mean of height

```python
meanHeight = df.height.mean()
meanHeight
```

```python
df.height = df.height.fillna(meanHeight)
df
```

#### Split Data in to X, Y

# X

```python
x = df.drop('Premium', axis=1)
x
```

# Y

```python
y = df['Premium']
y
```

#### Select the model

```python
model = linear_model.LinearRegression()
```

#### Fit the Model

```python
model.fit(x, y)
```

#### Model Coefficients or slops

```python
model.coef_
```

#### Model Intercept

```python
model.intercept_
```

#### Make Predictions

```python
model.predict([[27, 165.56, 60]])
```

```python
model.predict([[60, 165.10, 80]])
```

```python
predict = model.predict(x)
predict
```

```python
model.score(x, y)
```
